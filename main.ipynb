```python
import lightgbm as lgb
train_data = lgb.Dataset(X_train_scaled, label=y_train)
test_data = lgb.Dataset(X_test_scaled, label=y_test, reference=train_data)

params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'learning_rate': 0.1,
    'num_leaves': 31,
    'verbose': -1
}

model = lgb.train(params, train_data, valid_sets=[test_data], num_boost_round=100)

y_pred = model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Squared Error for LightGBM: {mse}')
print(f'Mean Absolute Error for LightGBM: {mae}')
mdae = median_absolute_error(y_test, y_pred)

print(f'Median Absolute Error for LightGBM: {mdae}')
```

```python
import pandas as pd



filtered_sales['delta_price'] = filtered_sales['price'].diff()
filtered_sales['delta_move'] = filtered_sales['move'].diff()

filtered_sales['price_change'] = filtered_sales['delta_price'] / filtered_sales['price'].shift(1)
filtered_sales['move_change'] = filtered_sales['delta_move'] / filtered_sales['move'].shift(1)

filtered_sales['elasticity'] = filtered_sales['move_change'] / filtered_sales['price_change']

print(filtered_sales['elasticity'].unique())
```
